```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 40
etha = 0.001
actions = []
time = []
for i in range(2, n+1):
    time.append(i)
    yse =  0.5
    yso = -0.5
    for j in range(2, i):
        if j % 2 == 0:
            yso += 1
            yse += -1
        else:
            yso += -1
            yse += 1
    if etha * yso + 1 < etha * yse + 1:
        actions.append(1)
    elif etha * yso + 1 > etha * yse + 1:
        actions.append(-1)
    else:
        rnumber = np.random.binomial(1, 0.5)
        if rnumber == 1:
            actions.append(1)
        else:
            actions.append(-1)

fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('Follow the Regularized Leader algorithm')
plt.plot(time, actions)
plt.show()
```
